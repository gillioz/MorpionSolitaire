# Morpion Solitaire examples

This notebook contains some examples on how to use the 
library MorpionSolitaire.py, including running a random exploration as well as importing/exporting files

```python
# import ipywidgets as widgets
```

```python
from MorpionSolitaire import *
```

```python
game = StartingGame()
```

```python
game.print()
```

```python
game2 = game.play(0)
game2.print()
```

```python
game3 = game2.play(-1)
game3.print()
```

```python
newgame = game.randomplay()
newgame.print()
```

```python
game.print()
```

```python
loopinggame = StartingGame()
for i in range(200):
    loopinggame = loopinggame.randomplay()
    if len(loopinggame.moves) == 0:
        break
loopinggame.print()
```

```python
loopinggame.grid.image()
```
