## Data generation

### Exploration algorithm

The algorithm that we are going to use is the following:
1. Play a game until the end using a given `model` (with a given `t`).
1. Rewind at random to an arbitrary intermediate step that we will call `base`.
1. From this intermediate step, explore the possible completions. The exploration is performed `n_completions` times according to the same `model` and `t` (and one time with `t = 0`). The number of steps `n` in each completion is recorded.
1. Add the the grid of `base` and statistics about the scores obtained in the exploration to the dataset. The algorithm saves the data (or returns it if no path is specified) if the form of:
    * an array of grids,
    * an array of scores, both corresponding to the base game,
    * an array containing the exploration data: mean, standard deviation, maximum, and first value of `n`.

Finally, repeat this loop as often as necessary to generate enough data. 

The parameters of the alogrithm are:
- `n_games`: total number of games to be played (i.e. size of the dataset)
- `n_completions`: for each game, how many random completions do we perform (higher number means more accurate exporation, but longer runtime)
- `folder`: path to the folder where the data should be saved (optional)
- `model` and `t`: arguments to be passed to the `play` method 
- `rules`: rules of the game, i.e. whether the initial game is of type 'cross' or 'pipe'
- `crop_grids`: whether to crop the grids' size before saving. This is used to avoid saving unnecessarily large grids that are mostly empty.


```python
def exploration_alg(n_games, path = None, rules = 'cross',
                    model = None, t = 10.0,
                    n_completions = 20, crop_grids = True):
    startinggame = StartingGame(rules)
    grids = []
    scores = []
    n_data = []
    for _ in trange(n_games):
        # 1. play at random
        base = startinggame.play(model = model, t = t, copy_grid = True)
        n = base.score
        # 2. rewind
        base = base.unplay(np.random.randint(base.score))
        n = [n - base.score]
        # 3. explore completions
        for _ in range(n_completions - 2):
            game = base.play(model = model, t = t, copy_grid = True)
            n.append(game.score - base.score)
        # play once with best predicted outcome
        game = base.play(model = model, t = 0.0, copy_grid = True)
        n.append(game.score - base.score)
        # 4. add result to the dataset
        grids.append(base.grid)
        scores.append(base.score)
        n_data.append([np.mean(n), np.std(n), np.max(n), n[0]])
    if crop_grids:
        non_zero_entries = np.nonzero(np.array(grids).sum((0,1)))
        imin = np.min(non_zero_entries)
        imax = np.max(non_zero_entries)
        grids = np.array(grids)[:,:,imin:imax,imin:imax]
        print('Grids cropped to size {0}x{0}'.format(imax - imin))
    if path is None:
        return (grids, scores, n_data)
    path = Path(path)
    if not path.exists():
        path.mkdir()
    np.save(path/'grids.npy', grids)
    np.save(path/'scores.npy', scores)
    np.save(path/'n.npy', n_data)
    print('Data saved to folder', path)
```

**Performance:** on my laptop, using a single CPU unit, the algorithm performs about 5 iterations per second, i.e. 1000 iterations in about 3 minutes, when there is no model (random exploration)

## Data analysis

Function that computes the frequency of scores in the data:

```python
def frequency(data, n_min = 0, n_max = None):
    if n_max is None:
        n_max = np.ceil(np.max(data)).astype(int)
    y, x = np.histogram(data, [i - 0.5 for i in range(n_min, n_max + 2)])
    x = x[:-1] + 0.5
    y = y/y.sum()
    return x, y
```

```python
def histogram(data, n_min = 0, n_max = None):
    x, y = frequency(data, n_min, n_max)
    plt.plot(x,y)
```

It is important to find an optimal 'temperature' parameter `t`:
- if `t` is too big, the exploration in random
- if `t` is too small, the exploration is limited to a neighborhood of the 'best' configuration

Therefore we introduce this function that performs a scan over different values of `t`, returning for each `t` the average and be result obtained:

```python
def t_scan(model, n_t, n_iterations, t_min = 0.1, t_max = 10.0):
    r = np.exp(np.log(t_min/t_max) / (n_t - 1))
    t = t_max
    t_list = []
    max_scores = []
    avg_scores = []
    batch_scores = []
    for i in trange(n_t * n_iterations):
        game = rules.play(model = model, t = t, copy_grid = True)
        batch_scores.append(game.score)
        if (i + 1) % n_iterations == 0:
            t_list.append(t)
            max_scores.append(np.max(batch_scores))
            avg_scores.append(np.mean(batch_scores))
            batch_scores = []
            t = r * t
    return t_list, avg_scores, max_scores
```
