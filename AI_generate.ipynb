## Exploration algorithm

```python
rules = StartingGame()
```

The algorithm that we are going to use is the following:
1. Play a game until the end using a given `model` and a given `temperature`
1. Rewind at random to an arbitrary intermediate step that we will call `exploration_game`
1. From this intermediate step, explore the possible completions and keep the best score. The exploration is performed `n_completions` times according to the same `model` and `temperature`, as well as one more time at zero temperature (i.e. choosing the 'best' completion) 
1. Add the the grid of `exploration_game` (x) and the maximal completion score (y) to the dataset.

Finally, repeat this loop as often as necessary to generate enough data. The algorithm returns a tuple that contains (in this order):
- an array containing the x data
- an array containing the y data
- an array of all scores attained in step 1
- an array of all scores corresponding to step 2
- an array of the maximum scores attained in step 3

The parameters of the alogrithm are:
- `n_games`: total number of games to be played (i.e. size of the dataset)
- `n_completions`: for each game, how many random completions do we perform (higher number means more accurate exporation, but longer runtime)
- `model` and `temperature`: arguments to be passed to the `play` method 


```python
def exploration_alg(n_games, n_completions = 20,
                    model = None, temperature = 10.0):
    data_grids = []
    data_n = []
    scores_rand = []
    scores_base = []
    scores_explore = []
    for _ in trange(n_games):
        # 1. play at random
        game = rules.play(model = model, t = temperature)
        scores_rand.append(game.score)
        # 2. rewind
        base = game.get_parent(np.random.randint(game.score))
        # 3. explore completions
        best_score = game.score
        # play once with best predicted outcome
        game = base.play(model = model, t = 0)
        if game.score > best_score:
            best_score = game.score
        # play many times with the given model
        for _ in range(n_completions - 2):
            game = base.play(model = model, t = temperature)
            if game.score > best_score:
                best_score = game.score
        # 4. add result to the dataset
        data_grids.append(base.grid.cell)
        data_n.append(best_score - base.score)
        scores_base.append(base.score)
        scores_explore.append(best_score)
    return (data_grids, data_n, scores_rand, scores_base, scores_explore)
```

**Performance:** on my laptop, using a single CPU unit, the algorithm performs about 5 iterations per second, i.e. 1000 iterations in about 3 minutes, when there is no model (random exploration)

In order to find the optimal 'temperature' for a run, it can be convenient to only run the first step of this algorithm and return the scores:

```python
def exploration_rand(n_games, model = None, temperature = 10.0):
    scores = []
    for _ in trange(n_games):
        game = rules.play(model = model, t = temperature)
        scores.append(game.score)
    return scores
```

Function that computes the frequency of scores in the data:

```python
def frequency(data):
    datamax = max(data)
    f = np.histogram(np.array(data), bins = datamax  + 1,
                     range = (-0.5, datamax + 0.5))[0]
    return f/f.sum()
```
