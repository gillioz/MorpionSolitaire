When I was younger, I used to play a pen-and-paper game called [*morpion solitaire*](https://en.wikipedia.org/wiki/Join_Five) (sometimes also *join five*, or simply the *line game*). 
The goal of the game is to draw as many lines and dots on a piece of paper following some simple rules.
To begin with, take a piece of graph paper and draw 36 dots in a cross pattern:

```python
#hide
grid = Grid('cross')
grid.print(view = 16)
plt.savefig('Part_1_Game_images/grid_empty.png', bbox_inches='tight')
plt.show()
```

![png](images/grid_empty.png 'The starting configuration of morpion solitaire.')

Then the rules say that you are allowed to add a point if you can draw a line that goes through that point and four other points that were there already. This is an example of a first move:

```python
#hide
grid.add_segment((10, 14, 1, 0))
grid.print(view = 16)
plt.savefig('Part_1_Game_images/grid_first_move.png', bbox_inches='tight')
plt.show()
```

![png](images/grid_first_move.png 'This is a valid first move.')

The lines can be horizontal, vertical, or diagonal, but always of lenght 4. The new point need not be at an extremity of the line, it can be anywhere along it. For instance, after the first move made above, it is possible to draw the diagonal line:

```python
#hide
grid.add_segment((12, 16, 3, 2))
grid.print(view = 16)
plt.savefig('Part_1_Game_images/grid_second_move.png', bbox_inches='tight')
plt.show()
```

![png](images/grid_second_move.png 'This is a second move only made possible after the first.')

The lines can cross, but they are not allowed to overlap. This means that the grid is slowly going to fill up, and at some point it might be that no more moves are possible. At this stage the game is over, and your score is how many points you have on the grid (not counting the initial 36 points).
Here is an animated gif showing a complete game:

```python
#hide
game = NewGame()
for i in range(35):
    game.print(view = 16);
    plt.savefig('Part_1_Game_images/grid_animation/grid_{}.png'.format(i), bbox_inches='tight')
    plt.close()
    game = game.play(index = 17, depth = 1)
```

![gif](images/grid_animation.gif 'A complete game, not particularly well played.')

The final score says 35. This is pretty low: if you try by yourself, I'm sure that you will reach much higher scores.
Playable versions of the game can be found as cell phone applications and all around the internet, for instance at:
- [joinfive.com](http://joinfive.com/)
